```python
import cv2
import glob


for img in glob.glob("./train/benign/*.jpg"):
    img_main = cv2.imread(img)
    grayScale = cv2.cvtColor( img_main, cv2.COLOR_RGB2GRAY )
    kernel = cv2.getStructuringElement(1,(17,17))
    blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)
    ret,thresh2 = cv2.threshold(blackhat,10,255,cv2.THRESH_BINARY)
    dst = cv2.inpaint(img_main,thresh2,1,cv2.INPAINT_TELEA)
    cv2.imwrite(img,dst)
    print(img) 
    
for img in glob.glob("./train/malignant/*.jpg"):
    img_main = cv2.imread(img)
    grayScale = cv2.cvtColor( img_main, cv2.COLOR_RGB2GRAY )
    kernel = cv2.getStructuringElement(1,(17,17))
    blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)
    ret,thresh2 = cv2.threshold(blackhat,10,255,cv2.THRESH_BINARY)
    dst = cv2.inpaint(img_main,thresh2,1,cv2.INPAINT_TELEA)
    cv2.imwrite(img,dst)
    print(img)
```
